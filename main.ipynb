```python
# 1. Import library
import cv2
from skimage.feature import graycomatrix, graycoprops
import matplotlib.pyplot as plt
```

```python
# 2. Load gambar dan tampilkan
img = cv2.imread('k1.jpg', 0)
plt.imshow(img, cmap='gray')
plt.title('Gambar Grayscale')
plt.axis('off')
plt.show()
```

```python
# 3. Hitung GLCM
glcm = graycomatrix(img, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
```

```python
# 4. Ambil fitur
print("Contrast:", graycoprops(glcm, 'contrast')[0, 0])
print("Energy:", graycoprops(glcm, 'energy')[0, 0])
print("Homogeneity:", graycoprops(glcm, 'homogeneity')[0, 0])
```
